Import Library

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

Import Data

```python
from sklearn.datasets import load_digits
```

```python
Sh=load_digits()
```

Data Preprocessing

```python
Sh.images.shape
```

```python
Sh.images[0]
```

```python
len(Sh.images)
```

```python
n_samples=len(Sh.images)
data=Sh.images.reshape((n_samples,-1))
```

```python
data [0]
```

```python
data[0].shape
```

```python
data.shape
```

Scaling Image Data

```python
data.min()
```

```python
data.max()
```

```python
data=data/16
```

```python
data.min()
```

```python
data.max()
```

```python
data [0]
```

Train Test Split Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(data,Sh.target, test_size=0.7,random_state=222529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

Random Forest Model

```python

from sklearn.ensemble import RandomForestClassifier
```

```python
pf=RandomForestClassifier()
```

```python
pf.fit(x_train,y_train)
```

Predict Test Data

```python
Y_pred = pf.predict(x_test)
```

```python
Y_pred
```

Model Acuracy

```python

from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,Y_pred))
```
